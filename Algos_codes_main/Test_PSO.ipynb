## PSO classique

```python
from PSO import *
# Example usage
if __name__ == "__main__":
    calls = 0
    def objective_function(x):
        global calls
        calls+=1
        return (x[0]-1)**2 + (x[1]-2)**2  # Objective function to minimize

    pso = PSO(objective_function, num_particles=4, dim=2, max_iterations=100, phi1=2.0, phi2=2.0, inertia_weight=0.5)
    best_position, best_value = pso.optimize()
    print("Global best position:", best_position)
    print("Global best value:", best_value)
    print(calls)
```

## PSO multi objectif pour obtenir le Front de Pareto

```python
import numpy as np
from PSO import *

# Example usage
if __name__ == "__main__":
    calls = 0

    def objective_function(x):
        global calls
        calls+=1
        return [(x[0]-1)**2 + (x[1]-2)**2, (x[0]+1)**2 + (x[1]-1)**2]  # Bi-objective function to minimize

    pso = MO_PSO(objective_function, num_particles=30, dim=2, max_iterations=100, phi1=2.0, phi2=2.0, inertia_weight=0.5, num_objectives=2)
    pareto_front = pso.optimize()
    print("Pareto Front:")
    for particle_position, particle_value in pareto_front:
        print("Position:", particle_position, "Values:", particle_value)
print(calls)
```

```python
pareto_values = [pareto_front[i][1] for i in range(len(pareto_front))]
```

```python
import matplotlib.pyplot as plt
x = [pareto_values[i][0] for i in range (len(pareto_values))]
y = [pareto_values[i][1] for i in range (len(pareto_values))]

plt.scatter(x,y)
```
